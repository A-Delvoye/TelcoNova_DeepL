Import et nettoyage des données

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("../Churn.csv")
df.head()
```

```python
doublons = df[df.duplicated()]
print(doublons)
```

```python
df.info()
df.describe()
```

```python
df["TotalCharges"] = df["TotalCharges"].replace(" ", pd.NA)
df.head()
```

suppression des manquants de totalcharges

```python
df = df.dropna(subset=["TotalCharges"])
df.head()
```

```python
n_missing = df["TotalCharges"].isna().sum()
print(f"TotalCharges manquants : {n_missing}")
print(df[df["TotalCharges"].isna()])
```

traitement des types des colonnes 

```python
df["TotalCharges"] = df["TotalCharges"].astype("float32")
```

Suppression des colonnes inutiles


```python
df.drop(columns=["customerID"], inplace=True)
```

Typage des colonnes 

```python
# Booléens
df["SeniorCitizen"] = df["SeniorCitizen"].astype("bool")
for col in ["Partner","Dependents","PhoneService","PaperlessBilling","Churn"]:
    df[col] = df[col].map({"Yes": True, "No": False})

# Catégorielles
cat_cols = [
    "gender","MultipleLines","InternetService","OnlineSecurity",
    "OnlineBackup","DeviceProtection","TechSupport","StreamingTV",
    "StreamingMovies","Contract","PaymentMethod"
]
for col in cat_cols:
    df[col] = df[col].astype("category")

# Numériques
df["tenure"] = df["tenure"].astype("int16")
df["MonthlyCharges"] = df["MonthlyCharges"].astype("float32")


```

# one hot encoding 


```python

df = pd.get_dummies(
    df,
    columns=cat_cols,
    drop_first=False,
    dtype=int
)

df.head()
df.info()
```

```python
df.head()
```

# conversion des bool en int

```python

bool_cols = df.select_dtypes(include="bool").columns


df[bool_cols] = df[bool_cols].astype("int8")


print(df[bool_cols].head())
print(df[bool_cols].dtypes)
```

# construction de l’echantillon d’entrainement et de test

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X = df.drop(columns=["Churn"])
y = df["Churn"].astype("int8")

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)

scaler = StandardScaler()
num_cols = ["tenure","MonthlyCharges","TotalCharges"]
X_train[num_cols] = scaler.fit_transform(X_train[num_cols])
X_test[num_cols]  = scaler.transform(X_test[num_cols])
```

Modele TensorflowKeras

```python
import tensorflow as tf

input_dim = X_train.shape[1]

def build_model():
    return tf.keras.Sequential([
        tf.keras.layers.Input(shape=(input_dim,)),
        tf.keras.layers.Dense(64, activation="relu"),
        tf.keras.layers.Dense(64, activation="relu"),
        tf.keras.layers.Dense(1, activation="sigmoid")
    ])

model = build_model()
model.compile(
    optimizer="adam",
    loss="binary_crossentropy",
    metrics=["accuracy"]
)

history = model.fit(
    X_train, y_train,
    validation_split=0.1,
    epochs=50,
    batch_size=32,
    verbose=1
)
```

# Evaluation du modele

```python
loss, acc = model.evaluate(X_test, y_test)
print(f"Test accuracy: {acc:.4f}")
```
